```python
import pandas as pd
import numpy as np
import re

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation

import pyLDAvis
from pyLDAvis import lda_model
```

```python
pyLDAvis.enable_notebook()
```

```python
pts_df = pd.read_csv('/Users/sambickel-barlow/Desktop/PP422/Final Project/tesco_parsed_processed.csv')
```

```python
pts_df['processed']
```

```python
pts_df['processed'] = pts_df['processed'].fillna("")
```

```python
# Convert text to a document-term matrix using TF-IDF
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(pts_df['processed'])
```

```python
# Fit LDA Model
n_topics = 25  # Adjust based on optimization
lda_model = LatentDirichletAllocation(n_components=n_topics, random_state=42)
lda_topics = lda_model.fit_transform(X)
```

```python
pyLDAvis.lda_model.prepare(
    lda_model = lda_model,
    dtm = X,
    vectorizer = vectorizer,
    # Keep our original sorting of documents
    sort_topics=False
)
```
